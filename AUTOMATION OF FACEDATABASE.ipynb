```python
import cv2
import numpy as np
import os
import scipy
import statistics
from datetime import datetime
import firebase_admin
from firebase_admin import credentials
from firebase_admin import db
cap = cv2.VideoCapture(0)
cap.set(3,640)
cap.set(4,480)

try:
    app = firebase_admin.get_app()
except ValueError as e:
    cred = credentials.Certificate("firebase_sdk.json")
    firebase_admin.initialize_app(cred, {
    'databaseURL': 'https://attendance-d27b4-default-rtdb.firebaseio.com/'})
    
    
Total_classes_conducted = 0
count = 0
sum_of_classes = [0]*len_of_folders
classes_to_attend_list = [0]*len_of_folders
att_percentage_list = [0]*len_of_folders
name_list = list(range(1, len_of_folders+1))
new_name_list=[str(x) for x in name_list]
strng2="ref"
filename_list = [strng2+ x for x in new_name_list]
strng3="/Person"
person_name = [strng3+x for x in new_name_list]
strng6 = "/Verified-Person"
verified_list = [strng6+x for x in new_name_list]
strng7 = "/Remarks"
remarks_list = [strng7+x for x in new_name_list]
# strng4 = "/TotalClassestoattend"
# total_classes_list = [strng4+x for x in new_name_list]
# strng5 = "/TotalClassesconducted"
# total_classes_conducted = [strng5+x for x in new_name_list]
print(filename_list)
for i in range(0,len(filename_list)):
    filename_list[i] = db.reference('/Person'+new_name_list[i])
    verified_list[i] = db.reference('/Verified-Person'+new_name_list[i])
    remarks_list[i] = db.reference('/Person'+new_name_list[i]+'/Remarks')
    #total_classes_conducted[i] = db.reference('/Person'+new_name_list[i]+"/TotalClassesConducted")
#     total_classes_list[i] = db.reference('/Person'+new_name_list[i]+"/TotalClassestoattend")
    
def find_classes(classes_present,total_classes):
    classes_to_attend=0
    att = classes_present/total_classes
    while(att<0.75):
        classes_present+=1
        total_classes+=1
        att=classes_present/total_classes
        classes_to_attend+=1
    return classes_to_attend
    
    





face_cascade_src =  "D:/jupyter projects ml/face recognition/cascade/haarcascade_frontalface_default.xml"
face_cascade = cv2.CascadeClassifier(face_cascade_src)

recognizer = cv2.face.LBPHFaceRecognizer_create()
recognizer.read("D:/jupyter projects ml/face recognition/Training12.yml")

net = cv2.dnn.readNet("D:/jupyter projects ml/face recognition/yolo/yolov3.weights", "D:/jupyter projects ml/face recognition/yolo/darknet-master/darknet-master/cfg/yolov3.cfg")
classes = []
with open("D:/jupyter projects ml/face recognition/yolo/darknet-master/darknet-master/data/coco.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]
layer_names = net.getLayerNames()
output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]

Attendance=[]

def Detector(img):
    conf = 1000
    pred=0
    #Attendance=[]
    coordinates = []
    mod=[]

    height, width, channels = img.shape
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    # Detecting objects
    blob = cv2.dnn.blobFromImage(img, 0.00392, (416, 416), (0, 0, 0), True, crop=False)

    net.setInput(blob)
    outs = net.forward(output_layers)

    # Showing informations on the screen
    class_ids = []
    confidences = []
    boxes = []
    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5:
                # Object detected
                center_x = int(detection[0] * width)
                center_y = int(detection[1] * height)
                w = int(detection[2] * width)
                h = int(detection[3] * height)

                # Rectangle coordinates
                x = int(center_x - w / 2)
                y = int(center_y - h / 2)

                boxes.append([x, y, w, h])
                confidences.append(float(confidence))
                class_ids.append(class_id)

    indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)
    font = cv2.FONT_HERSHEY_PLAIN
    for i in range(len(boxes)):
        if i in indexes:
            x, y, w, h = boxes[i]
            label = str(classes[class_ids[i]])
            if label == "person":
                coordinates.append(boxes[i])
                # cv2.rectangle(img, (x, y), (x + w, y + h), color, 2)
                # cv2.putText(img, label, (x, y + 30), font, 3, color, 3)
    preDiction = []
    conFidence = []
    for i in coordinates:
        # print(i)
        # print(count)
        # cv2.rectangle(img, (i[0],i[1]), (i[0]+i[2],i[1]+i[3]), (255,0,0),1)
        # print(i)
        if(i[0]>=0 and i[1]>=0 and i[2]>=0 and i[3]>=0):
            person_segmented = gray[i[1]:i[1] + i[3], i[0]:i[0] + i[2]]

            faces = face_cascade.detectMultiScale(person_segmented, 1.1, 1)
            preDiction = []
            conFidence = []
            for (a,b,c,d) in faces:
                if c*d > 500:
                    face = person_segmented[b:b+d, a:a+c]
                    #cv2.imshow("Face", face)
                    #print(face.shape)
                    #face.resize(150,150)
                    pred, conf = recognizer.predict(gray)
                    preDiction.append(pred)
                    conFidence.append(conf)
           # if conf>40:
                #(str(pred) + "   " + str(conf))
                #cv2.imshow("Person", face)
                #for o in range(0,11):
                #Attendance.append(str(pred))
                #Attendance.append(pred)
            #mod=statistics.mode(Attendance)
#print(mod[0])

    return(preDiction, conFidence)
    #return(statistics.mode(Attendance))
        #print(Attendance)



while True:
    time.sleep(2)
    for i in range(0,len(filename_list)):
        filename_list[i].set({"Attendance":"0"})
    Total_classes_conducted+=1
    
#for i in range (0,10):
   # ret, img = cap.read()
    #cv2.imshow("Webcam Frame", img)
    #cv2.imwrite(str(id_person)+"_"+str(i)+".png",img)
    
    #cv2.waitKey(1)
#cap.release()
#cv2.destroyAllWindows()
    #print(pred)
    ret, frame = cap.read()
    emptyList = []
    x, y=Detector(frame)
    print(x)
    print(y)
    count+=1
    for j in range (0,len(y)):
        
        print(count)

        if y[j]>30:
            (filename_list[name_list.index(x[j])]).set({"Attendance":"1"})
            sum_of_classes[name_list.index(x[j])] += (int)(filename_list[name_list.index(x[j])].get("/Attendance")[0]["Attendance"])
            verified_list[name_list.index(x[j])].set({"Total":(str)(sum_of_classes[name_list.index(x[j])])})
            att_percentage_list[name_list.index(x[j])] = (sum_of_classes[name_list.index(x[j])]/Total_classes_conducted)*100
            if(att_percentage_list[name_list.index(x[j])]>=75):
                remarks_list[name_list.index(x[j])].set({"Remarks":"Above 75 and current attendace is {0:.2f}%".format(att_percentage_list[name_list.index(x[j])])})
            else:
                remarks_list[name_list.index(x[j])].set({"Remarks":"Below 75 and current attendace is {0:.2f}%".format(att_percentage_list[name_list.index(x[j])])})
#             #print(sum_of_classes)
#             classes_to_attend_list[name_list.index(x[j])] = find_classes(sum_of_classes[name_list.index(x[j])],count)
# #             print(classes_to_attend_list)
#             total_classes_list[name_list.index(x[j])].set({"classes":(str)(classes_to_attend_list[name_list.index(x[j])])})
            #total_classes_conducted[x[j]-1].set({"TotalClassesConducted":(str)(count)})
        elif y[j]<=30:
            (filename_list[name_list.index(x[j])]).set({"Attendance":"0"})


    
        
   # print(Detector(frame))
    cv2.imshow("Frame", frame)
    key = cv2.waitKey(1) & 0xFF

    if key == 27:
        break
# print(Attendance)        
# #print(Detector(frame))
# print("attendance marked for roll no: ",statistics.mode(Attendance))
cap.release()
cv2.destroyAllWindows()
#print("attendance marked for,roll no",statistics.mode(Attendance))
```
