```python
import os
import cv2
import numpy as np
from PIL import Image

face_cascade_src = 'D:/jupyter projects ml/face recognition/cascade/haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier(face_cascade_src)

names = []
paths = []

for users in os.listdir("D:/jupyter projects ml/face recognition/face dataset"):
    
#     if (users.endswith(".ipynb")):
#         print("Skipping " + users)
#     else:
    names.append(users)


for name in names:
    for image in os.listdir("D:/jupyter projects ml/face recognition/face dataset/{}".format(name)):
        pathString = os.path.join("D:/jupyter projects ml/face recognition/face dataset/{}".format(name), image)
        paths.append(pathString)

def find_max(faces):
    maxx = 0
    maxy = 0
    maxw = 0
    maxh = 0
    maxarea=0
    for (x,y,w,h) in faces:
        if (w*h > maxarea):
            maxarea = w*h
            maxx = x
            maxy = y
            maxw = w
            maxh = h
    return (maxx, maxy, maxw, maxh)


faces_to_train = []
ids = []

for img_path in paths:
    img = Image.open(img_path).convert("L")
    img = np.array(img, "uint8")
    print(type(img))
    faces = face_cascade.detectMultiScale(img, 1.1, 1)
    
    # cv2.imshow(img_path,img)

    #img_np = np.array(img, "uint8")

    #faces.append(img_np)
    #faces = [[ 40, 156 , 30 , 30],
 #[246 , 22 ,169 ,169],
 #[277, 239 ,105, 105],
 #[174, 152, 215, 215],
 #[ 92 ,176, 30 , 30]]
    #(a,b,c,d)= find_max(faces)
    #print((a,b,c,d))
    #find_max(faces)

    for (a,b,c,d) in faces:
        (a,b,c,d)= find_max(faces)
        if c*d > 100:
            face = img[b:b+d, a:a+c]
            cv2.imshow("Face", face)
            img_numpy = np.array(face, "uint8")
            #faces.append(face)
            #print(type(face))
            faces_to_train.append(face)
            break
            #print(face.shape)

    id = int(img_path.split("/")[4].split("\\")[0].split("_")[0])
    #print(id)

    ids.append(id)

ids = np.array(ids)
print(ids)

cv2.waitKey(0)
cv2.destroyAllWindows()

print(names)
# print(paths)
# print(faces)

trainer = cv2.face.LBPHFaceRecognizer_create()
trainer.train(faces_to_train,ids)
trainer.write("D:/jupyter projects ml/face recognition/Training12.yml")
```
